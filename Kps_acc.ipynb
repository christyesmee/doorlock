```python
import csv
import random
import numpy as np

# Define the layout of the keypad with coordinates (x, y) relative to the accelerometer
keypad_layout = {
    '1': (-1, 1), '2': (0, 1), '3': (1, 1),
    '4': (-1, 0), '5': (0, 0), '6': (1, 0),
    '7': (-1, -1), '8': (0, -1), '9': (1, -1),
    '0': (0, -2)
}

# Function to calculate the realistic time interval between two keys
def calculate_realistic_distance(key1, key2, keypad_layout, key_spacing_cm, speed_cm_per_sec, min_press_time_sec):
    if key1 == key2:
        # If pressing the same key, return the minimum press time
        return min_press_time_sec
    else:
        # Calculate the Euclidean distance between the keys
        pos1 = keypad_layout[key1]
        pos2 = keypad_layout[key2]
        distance_cm = np.sqrt((pos1[0] - pos2[0])**2 + (pos1[1] - pos2[1])**2) * key_spacing_cm
        # Calculate the time interval based on distance and speed
        time_sec = distance_cm / speed_cm_per_sec
        return time_sec

# Variables for the keypad and speed
key_spacing_cm = 1.5  # Spacing between keys in cm
speed_cm_per_sec = 5   # Speed of key press in cm/sec
min_press_time_sec = 0.2  # Minimum time for pressing a key in sec

# Generate realistic intervals
def generate_realistic_intervals(num_samples, keypad_layout, key_spacing_cm, speed_cm_per_sec, min_press_time_sec):
    intervals = []
    all_keys = list(keypad_layout.keys())

    for _ in range(num_samples):
        # Randomly select 4 keys to simulate a password entry sequence
        password_keys = random.sample(all_keys, 4)

        # Calculate the time intervals between consecutive key presses
        time_intervals = [
            calculate_realistic_distance(password_keys[i], password_keys[i+1], keypad_layout, key_spacing_cm, speed_cm_per_sec, min_press_time_sec)
            for i in range(3)
        ]

        # Add the time intervals to the list
        intervals.append(time_intervals)

    return intervals

realistic_intervals = generate_realistic_intervals(100, keypad_layout, key_spacing_cm, speed_cm_per_sec, min_press_time_sec)

# Calculate cumulative times for the intervals
cumulative_times = [[sum(interval_set[:i+1]) for i in range(len(interval_set))] for interval_set in realistic_intervals]

# Combine cumulative times with original intervals for the final dataset
final_dataset = [cumulative + original for cumulative, original in zip(cumulative_times, realistic_intervals)]

# Define header for the CSV file
headers = ['cum time 1', 'cum time 2', 'cum time 3', 'int 1', 'int 2', 'int 3']

# Write the dataset to a CSV file
csv_filename = '/content/drive/My Drive/Internet of Things Security/combined_dataset.csv'
with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(headers)  # Write the header
    writer.writerows(final_dataset)  # Write the data

# Display the first few rows of the final dataset as a sample
final_dataset[:5]
```
